## Further options and ways to save the plot image

Let's use an example for cats and show some of the options and how to save the plot.

````r
cats<-showPhylo(speciesNames=c("puma","leopard","jaguar","domestic cat"), nameType="common")
````

````r
plot(cats)
````

We can save that file as an image using `ggsave` to specify output to a file name and extension of choice.

````r
ggsave("cats_tree.jpg")
````

That will save a file named `cats_tree.jpg`.If you look now in the file navigator pane to the left, you'll see it listed shortly. You can double-click on it to open.

Prefer a Portable Network Graphics (`.png`) formatted image instead? Try running this cell:

````r
ggsave("cats_tree.png")
````

That should produce a file with a `.png` extension in the file file navigation panel to the left.
